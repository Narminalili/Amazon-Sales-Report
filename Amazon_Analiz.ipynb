# Dimensionality Reduction: Pruning Irrelevant Features

```python
df.drop(['Unnamed: 22', 'fulfilled-by','ship-country', 'index'], axis=1, inplace=True)
df.columns
```

# How many null values are there in total in each column?

```python
df.isnull().sum()
```

# Records with missing Amount values were removed to maintain financial data integrity and ensure analysis accuracy.

```python
df.dropna(subset=['Amount'], inplace=True)
```

# Remaining missing values in categorical columns were standardized to ensure dataset completeness.

```python
df['promotion-ids'] = df['promotion-ids'].fillna('No Promotion')
df['Courier Status'] = df['Courier Status'].fillna('Unknown')
```

# Missing values in location-based columns were handled to ensure consistency across the dataset.

```python
for col in ['ship-city', 'ship-state', 'ship-postal-code']:
    df[col] = df[col].fillna('Unknown')
```

# A final check was performed to ensure that all missing values were successfully addressed.

```python
df.isnull().sum()
```

# Data Types Verification

```python
df.info()
```

# The `Date` column was converted to a standardized datetime format to enable time-series analysis.

```python
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%y', errors='coerce')
```

# To ensure consistent formatting and prevent mathematical errors, the `postal code` field was standardized as a string.

```python
df['ship-postal-code'] = pd.to_numeric(df['ship-postal-code'], errors='coerce').fillna(0).astype(int)
```

```python
df.to_csv('Amazon_Sale_Report_Cleaned.csv', index=False)
```

# How has the total revenue evolved over the specified time period?

```python
daily_revenue = df.groupby('Date')['Amount'].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(data=daily_revenue, x='Date', y='Amount', color='darkblue', linewidth=2, marker='o')
plt.title('Total Revenue Evolution Over Time', fontsize=15, fontweight='bold')
plt.xlabel('Date', fontsize=12)
plt.ylabel('Total Revenue (INR)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

# What is the distribution of product categories in the orders?

```python
category_sales = df.groupby('Category')['Amount'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x=category_sales.index, y=category_sales.values, hue=category_sales.index, palette='magma', legend=False)

plt.title('Revenue by Category', fontsize=15)
plt.xlabel('Product Category')
plt.ylabel('Total Sales Amount')
plt.xticks(rotation=45)
plt.show()
```

# What is the typical quantity of products per order?

```python
plt.figure(figsize=(10, 6))

sns.countplot(data=df, x='Qty', hue='Qty', palette='coolwarm', legend=False)

plt.title('Distribution of Product Quantity per Order', fontsize=15, fontweight='bold')
plt.xlabel('Quantity per Order', fontsize=12)
plt.ylabel('Order Count', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
print("Quantity Statistics:")
print(df['Qty'].describe())
```

# Create a heatmap to visualize the distribution of sales across months and days of the week based on the dataset.

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month_name()
df['Weekday'] = df['Date'].dt.day_name()

days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
months_order = df['Month'].unique()

heatmap_data = df.pivot_table(index='Month', columns='Weekday', values='Amount', aggfunc='sum')
heatmap_data = heatmap_data[days_order] 

plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, annot=True, fmt=".0f", cmap="YlGnBu", linewidths=.5)

plt.title('Sales Heatmap: Months vs. Weekdays', fontsize=15, fontweight='bold')
plt.xlabel('Day of the Week', fontsize=12)
plt.ylabel('Month', fontsize=12)

plt.show()
```
